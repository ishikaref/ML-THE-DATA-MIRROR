**40.** We have a hypothesis that as the day goes on, the tips get higher.  We'll need to wrangle the data a bit before testing this, however.  First, now that we have a datetime column, pickup_time, create a subset of it to create a new DataFrame, taxis_one_day. This new DataFrame should have values between '2019-03-23 00:06:00' (inclusive) and '2019-03-24 00:00:00' (exlusive).

````python
import pandas as pd
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/taxis.csv"
taxis = pd.read_csv(url)
taxis["pickup_time"] = pd.to_datetime(taxis["pickup"])
start_time = pd.to_datetime("2019-03-23 00:06:00")
end_time   = pd.to_datetime("2019-03-24 00:00:00")
taxis_one_day = taxis[(taxis["pickup_time"] >= start_time) &
                      (taxis["pickup_time"] < end_time)]
print(taxis_one_day.head())
print("\nNumber of rows in taxis_one_day:", len(taxis_one_day))
````

**41.** We now have a range from morning until midnight, but we to take the mean of the numeric columns, grouped at one hour intervals.  Save the result as df_means, and display it.

````python
import pandas as pd
taxis_one_day["pickup_hour"] = taxis_one_day["pickup_time"].dt.hour
df_means = taxis_one_day.groupby("pickup_hour").mean(numeric_only=True)
print(df_means)
````

**42.** Create a simple line plot of the value "distance".  

````python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/taxis.csv"
taxis = pd.read_csv(url)

taxis["pickup_time"] = pd.to_datetime(taxis["pickup"])

plt.plot(taxis.index, taxis["distance"], color="blue", linestyle="-", marker="o")

plt.xlabel("Index")
plt.ylabel("Distance")
plt.title("Line Plot of Taxi Trip Distances")
plt.grid(True)
plt.show()
````

**43.** Overall, do riders travel further or less far as the day progresses?

````python
import pandas as pd
taxis['pickup'] = pd.to_datetime(taxis['pickup'])
taxis['hour'] = taxis['pickup'].dt.hour
avg_distance_by_hour = taxis.groupby('hour')['distance'].mean()
print(avg_distance_by_hour)
import seaborn as sb
import matplotlib.pyplot as plt

sb.lineplot(x=avg_distance_by_hour.index, y=avg_distance_by_hour.values)
plt.xlabel("Hour of Day")
plt.ylabel("Average Trip Distance (miles)")
plt.title("Taxi Trip Distance vs. Time of Day")
plt.show()
````

So overall, trip distances tend to increase later in the day, especially in the evening when leisure and airport travel are more common.

**47.** Admittedly, the size of the data set is fairly small given how we've subsetted it.  But based on the values in #45, which of the three pairs show the strongest correlation.

````python
correlations = {
    "tip_in_percent vs distance": df_means["tip_in_percent"].corr(df_means["distance"]),
    "tip_in_percent vs passengers": df_means["tip_in_percent"].corr(df_means["passengers"]),
    "tip_in_percent vs time_interval": df_means["tip_in_percent"].corr(df_means["time_interval"])
}

print(correlations)

strongest_pair = max(correlations, key=lambda k: abs(correlations[k]))
print("\nStrongest correlation is:", strongest_pair, "=", correlations[strongest_pair])
````
